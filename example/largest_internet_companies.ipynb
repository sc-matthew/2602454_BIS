```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

import ipywidgets as widgets
from ipywidgets import interact
```

```python
# list of DataFrames
dfs=pd.read_html('https://en.wikipedia.org/wiki/List_of_largest_Internet_companies')
len(dfs)
```

```python
df=dfs[1]
df
```

```python
df.columns
```

```python
df['Revenue ($B)']=pd.to_numeric(df['Revenue ($B)'].str.replace('$', ''))
df
```

```python
df['Market cap. ($B)']=pd.to_numeric(df['Market cap. ($B)'].str.replace(r'[\$,\-]', '', regex=True))
df
```

```python
# df['Market cap. ($B)']=pd.to_numeric(df['Market cap. ($B)'].str.replace('$', '').str.replace(',', '').str.replace('-', ''))
df
```

```python
# df['Market cap. ($B)']=pd.to_numeric(df['Market cap. ($B)'].str.replace('$', '').str.replace(',', '').str.replace('-', ''))
# df
```

```python
df['Employees']=pd.to_numeric(df['Employees'].str.replace('(2019)', '', regex=False).str.replace(',', '', regex=False))
df
```

```python
df.drop(columns=['Refs'], inplace=True)
df
```

```python
df.info()
```

## Interactive DataFrame


```python
# decorator
# python decorator
```

```python
@interact
def filter(regex_pat="."):
    return df[(df["Company"].str.contains(regex_pat, regex=True, case=False, na=False))].sort_values(
        "Company", ascending=True, key=lambda v: v.str.lower()
    )

```

```python
@interact
def filter(regex_pat="."):
    return df[(df["Industry"].str.contains(regex_pat, regex=True, case=False, na=False))].sort_values(
        "Company", ascending=True
    )

```
