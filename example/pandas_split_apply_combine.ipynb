# Understand `S`plit-`A`pply-`C`ombine

```python
import pandas as pd
print(f'pandas version: {pd.__version__}')
pd.Timestamp.now()
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df.sample(5)
```

```python
df=df[df.year==2008]
df.sample(5)
```

```python
df.shape
```

```python
df['cty'].mean()
```

```python
g=df.groupby('class')
g.groups
```

```python
g.groups.keys()
```

```python
g.groups['compact']
```

```python
g.get_group('compact').head()
```

```python
for k in g.groups.keys():
    print(f"{k}: {g.get_group(k)['cty'].mean()}")
```

```python
g['cty'].mean()
```

```python
g[['cty', 'hwy']].mean()
```

## split-apply-combine

### nlargest

```python
df.nlargest(1, 'cty')
```

```python
g=df.groupby('class')
```

```python
g.groups
```

```python
g.get_group('compact').nlargest(3, 'cty')
```

```python
g.get_group('suv').nlargest(3, 'cty')
```

```python
# g.nlargest(3, 'cty') # -> error
```

```python
g.apply(lambda grp: grp.nlargest(1, 'cty'))
```

```python
g.apply(lambda grp: grp.nlargest(1, 'cty')).sort_values('cty', ascending=False)
```

## multilevel group

```python
g2=df.groupby(['class', 'drv'])
g2.groups
```

```python
g2[['cty', 'hwy']].mean()
```

```python
g2.describe()[['cty', 'hwy']]
```

## rank within group

```python
g=df.groupby('class')
```

```python
df['rank'] = g['cty'].rank('dense', ascending=False)
```

```python
df[df['class']=='compact'].sort_values('rank')
```

```python
df.sort_values(by=['class', 'rank'])
```

```python
g.apply(lambda grp: grp.sort_values(by=['rank']))
```
